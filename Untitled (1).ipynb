```python
def rect_spherical(lat, lon):
    
    rad_lat = np.radians(lat)
    rad_lon = np.radians(lon)
    
    x1 = np.cos(rad_lat) * np.cos(rad_lon)
    x2 = np.cos(rad_lat) * np.sin(rad_lon)
    x3 = np.sin(rad_lat)
    
    return np.array([x1, x2, x3])

def rotation(R, x):
    
    xp = np.dot(R, x)
    
    return xp

def primed_angles(xp):
    
    longp = np.arctan(xp[1], xp[0])
    latp = np.arcsin(xp[2])
    
    return longp, latp

def RADEC_HADEC(LST, radian = 'None'):
    
    if radian == 'None':
        
        rad_lst = np.radians(LST)
        R = np.array([[np.cos(rad_lst), np.sin(rad_lst), 0], [np.sin(rad_lst), -np.cos(rad_lst), 0], [0,0,1]])

        return R
    
    if radian == 'True':
        R = np.array([[np.cos(rad_lst), np.sin(rad_lst), 0], [np.sin(rad_lst), -np.cos(rad_lst), 0], [0,0,1]])

        return R
        

def HADEC_AZALT(lat, radian = 'None'):
    if radian == 'None':
        
        rad_lat = np.radians(lat)
        R = np.array([[-np.sin(rad_lat), 0, np.cos(rad_lat)], [0, -1, 0], [np.cos(rad_lat),0,np.sin(rad_lat)]])

        return R
    
    else:
        R = np.array([[-np.sin(rad_lat), 0, np.cos(rad_lat)], [0, -1, 0], [np.cos(rad_lat),0,np.sin(rad_lat)]])

        return R
        
    
def EQ_GALACTIC():
    
    R = np.array([[-.054876, -.873437, -.483835],[.494109, -.444830,.746982],[-.867666, -.198076, .455984]])
    return R

```

```python
def load_data(filename):
    '''
    Function that loads in the data
    
    '''
    data = np.load(filename)

    return data

def scale(data, volt):
    
    '''
    Function that scales the data that we have to the proper scaling
    '''
    
    scaled = (data*volt)/(2**16)
    
    return scaled

def reshape_data(data):
    
    '''
    Reshaping the data so that we have a 2-D array one for imaginary and real
    '''

    new_data = data.reshape(2, len(data)//2)

    return new_data

def splitting_data(data):

    samples = len(data[0])//16000

    real = np.split(data[0], samples)
    imag = np.split(data[1], samples)

    return real, imag

def complex(real, imag):

    comp = []

    for i in range(len(real)):
        comp_val = real[i] + 1j*imag[i]

        comp.append(comp_val)

    return np.array(comp)

def frequency(vsamp):

    data_size = 16000
    dat = np.fft.fftfreq(data_size, d = 1/vsamp)
    
    freq = dat/1e6
    return freq

def power_transform(data):

    pow = []

    for i in data:

        fft = np.fft.fft(i)
        power = np.abs(fft)**2
        pow.append(power)

    return np.array(pow)

def avg_power(power):

    mean_power = np.mean(power, axis = 0)

    return mean_power

def median_power(power):

    median = np.median(power, axis = 0)

    return median


def avg_filt(data, freq, bins):

    if bins == 5:

        smoothed_data = []
        smoothed_freq = []

        num_split = len(data)//5

        split_data = np.split(data, num_split)
        split_freq = np.split(freq, num_split)

        for i in range(len(split_data)):

            mean_data = np.mean(split_data[i], axis = 0)
            mean_freq = np.mean(split_freq[i], axis = 0)

            smoothed_data.append(mean_data)
            smoothed_freq.append(mean_freq)


        return np.array(smoothed_freq), np.array(smoothed_data)

    if bins == 10:

        smoothed_data = []
        smoothed_freq = []

        num_split = len(data)//10

        split_data = np.split(data, num_split)
        split_freq = np.split(freq, num_split)

        for i in range(len(split_data)):

            mean_data = np.mean(split_data[i], axis = 0)
            mean_freq = np.mean(split_freq[i], axis = 0)

            smoothed_data.append(mean_data)
            smoothed_freq.append(mean_freq)


        return np.array(smoothed_freq), np.array(smoothed_data)
    
    if bins == 20:

        smoothed_data = []
        smoothed_freq = []

        num_split = len(data)//20

        split_data = np.split(data, num_split)
        split_freq = np.split(freq, num_split)

        for i in range(len(split_data)):

            mean_data = np.mean(split_data[i], axis = 0)
            mean_freq = np.mean(split_freq[i], axis = 0)

            smoothed_data.append(mean_data)
            smoothed_freq.append(mean_freq)


        return np.array(smoothed_freq), np.array(smoothed_data)

def median_filt(data, freq, bins):

    if bins == 5:

        smoothed_data = []
        smoothed_freq = []

        num_split = len(data)//5

        split_data = np.split(data, num_split)
        split_freq = np.split(freq, num_split)

        for i in range(len(split_data)):

            med_data = np.median(split_data[i], axis = 0)
            med_freq = np.median(split_freq[i], axis = 0)

            smoothed_data.append(med_data)
            smoothed_freq.append(med_freq)


        return np.array(smoothed_freq), np.array(smoothed_data)

    if bins == 7:

        smoothed_data = []
        smoothed_freq = []

        num_split = len(data)//10

        split_data = np.split(data, num_split)
        split_freq = np.split(freq, num_split)

        for i in range(len(split_data)):

            med_data = np.median(split_data[i], axis = 0)
            med_freq = np.median(split_freq[i], axis = 0)

            smoothed_data.append(med_data)
            smoothed_freq.append(med_freq)


        return np.array(smoothed_freq), np.array(smoothed_data)
    
    if bins == 9:

        smoothed_data = []
        smoothed_freq = []

        num_split = len(data)//20

        split_data = np.split(data, num_split)
        split_freq = np.split(freq, num_split)

        for i in range(len(split_data)):

            med_data = np.median(split_data[i], axis = 0)
            med_freq = np.median(split_freq[i], axis = 0)

            smoothed_data.append(med_data)
            smoothed_freq.append(med_freq)


        return np.array(smoothed_freq), np.array(smoothed_data)
    
def Gain(s_cal, s_cold):

    T_cal = 300
    T_cold = 2.73

    diff = s_cal-s_cold

    G = ((T_cal - T_cold)/(np.sum(diff))) * np.sum(s_cold)

    return G

def T_power(line, G):

    power = line * G

    return power


def velocity(data):
    pass
    
```

```python
def f_x(a, freq):
    
    c = 3e8 #m/s
    length_fs = (c / (freq * 10**9)) * 100 #3.75 cm
    
    #print(length_fs)
    
    denom = 1-(length_fs/(2*a))**2
    
    
    #print(denom)
    sqrt = np.sqrt(denom)
    
    length = (length_fs / sqrt) 
    
    return length

a_vals = np.linspace(2.2, 2.5, 1000)

y = f_x(a_vals, 8)

#plt.plot(a_vals, y)
#plt.show()
wavelength = np.array([f_x(a, 8) for a in a_vals])

diff = abs((2*(x_stop[2]-x_stop[1])) - wavelength)

index_min = np.argmin(diff)

print('The difference between theory and observation is') 
print'minimized for the X_Band when a is' 
print'close to the value of: ', round(a_vals[index_min], 3), 'cm'
```
